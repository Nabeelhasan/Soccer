```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import urllib.request
```

```python
Ligue1_assists_url = "https://www.foxsports.com/soccer/stats?competition=43&season=20180&category=standard&sort=4"
```

```python
uClient = uReq(Ligue1_assists_url)
Ligue1_assists_url_page_html = uClient.read()
uClient.close()
```

```python
Ligue1_assists_page_soup = soup(Ligue1_assists_url_page_html,"html.parser")
```

```python
assist_Ligue1_playerName = Ligue1_assists_page_soup.findAll("a",{"class":"wisbb_fullPlayer"})
print(assist_Ligue1_playerName[0].span.text)
#a,b = assist_Ligue1_playerName[14].span.text.split(',')
```

```python
assist_Ligue1_Name = []
for i in range(50):
    get_assist_playerName = assist_Ligue1_playerName[i].span.text.split(', ')
    #print(T_get_assist_playerName)
    #a,b = T_get_assist_playerName.split(',')
    #if len(b) == 0
    if len(get_assist_playerName) == 2:
        a = get_assist_playerName[1]
        b = get_assist_playerName[0]
        c = a+' '+b
    else:
        b = get_assist_playerName[0]
        a = ''
        c = b
    assist_Ligue1_Name.append(c)
    
print(assist_Ligue1_Name)
    
    
```

```python
Ligue1_assistsPlusGoal_made = Ligue1_assists_page_soup.findAll('tr',{"class":"wisbb_fvStand"})
print(Ligue1_assistsPlusGoal_made[1].contents[9].text)
```

```python
Ligue1_assistsPlusGoal_made_list = []
for i in range(1,51):
    Ligue1_assists = Ligue1_assistsPlusGoal_made[i].contents[9].text
    Ligue1_assistsPlusGoal_made_list.append(Ligue1_assists)
    
print(Ligue1_assistsPlusGoal_made_list)
```

```python
Ligue1_assists_made = Ligue1_assists_page_soup.findAll('td',{"class":"wisbb_priorityColumn wisbb_selected"})
print(len(Ligue1_assists_made))
```

```python
Ligue1_assists_made_list = []
for i in range(50):
    Ligue1_assists = Ligue1_assists_made[i].text
    Ligue1_assists_made_list.append(Ligue1_assists)
    
print(Ligue1_assists_made_list)
```

```python
Ligue1_assist_club = Ligue1_assists_page_soup.findAll('span',{"class":"wisbb_tableAbbrevLink"})
print(Ligue1_assist_club[8].text)
print(Ligue1_assist_club[1].text)
print("asd")
```

```python
Ligue1_assist_club_list =[]
for i in range(50):
    get_club_name = Ligue1_assist_club[i].text
    #print( get_club_name)
    get_club_name = get_club_name[1:4]
    Ligue1_assist_club_list.append(get_club_name)
    
    #Ligue1_assist_club_list.append(get_club_prefix)
    
print(Ligue1_assist_club_list)
```

```python
PlayerID_Ligue1_assist_list = []
for i in range(50):
    make_id = assist_Ligue1_Name[i][0:4] + Ligue1_assist_club_list[i]
    PlayerID_Ligue1_assist_list.append(make_id)
    
print(PlayerID_Ligue1_assist_list)
```

```python
Ligue1_assist_data = list(zip(assist_Ligue1_Name,Ligue1_assist_club_list,Ligue1_assists_made_list,PlayerID_Ligue1_assist_list))
Ligue1_assist_df = pd.DataFrame(data = Ligue1_assist_data,columns = ['Player Name','Club','Assists','PlayerID'])
Ligue1_assist_df.head()
```
