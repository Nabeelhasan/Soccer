```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import urllib.request
```

```python
SerieA_assists_url = "https://www.foxsports.com/soccer/stats?competition=3&season=20180&category=STANDARD&pos=0&team=0&isOpp=0&sort=4&sortOrder=0&page=2"
```

```python
uClient = uReq(SerieA_assists_url)
SerieA_assists_url_page_html = uClient.read()
uClient.close()
```

```python
SerieA_assists_page_soup = soup(SerieA_assists_url_page_html,"html.parser")
```

```python
assist_SerieA_playerName = SerieA_assists_page_soup.findAll("a",{"class":"wisbb_fullPlayer"})
print(assist_SerieA_playerName[0].span.text)
#a,b = assist_SerieA_playerName[14].span.text.split(',')
```

```python
assist_SerieA_Name = []
for i in range(50):
    get_assist_playerName = assist_SerieA_playerName[i].span.text.split(', ')
    #print(T_get_assist_playerName)
    #a,b = T_get_assist_playerName.split(',')
    #if len(b) == 0
    if len(get_assist_playerName) == 2:
        a = get_assist_playerName[1]
        b = get_assist_playerName[0]
        c = a+' '+b
    else:
        b = get_assist_playerName[0]
        a = ''
        c = b
    assist_SerieA_Name.append(c)
    
print(assist_SerieA_Name)
    
    
```

```python
SerieA_assistsPlusGoal_made = SerieA_assists_page_soup.findAll('tr',{"class":"wisbb_fvStand"})
print(SerieA_assistsPlusGoal_made[1].contents[9].text)
```

```python
SerieA_assistsPlusGoal_made_list = []
for i in range(1,51):
    SerieA_assists = SerieA_assistsPlusGoal_made[i].contents[9].text
    SerieA_assistsPlusGoal_made_list.append(SerieA_assists)
    
print(SerieA_assistsPlusGoal_made_list)
```

```python
SerieA_assists_made = SerieA_assists_page_soup.findAll('td',{"class":"wisbb_priorityColumn wisbb_selected"})
print(len(SerieA_assists_made))
```

```python
SerieA_assists_made_list = []
for i in range(50):
    SerieA_assists = SerieA_assists_made[i].text
    SerieA_assists_made_list.append(SerieA_assists)
    
print(SerieA_assists_made_list)
```

```python
SerieA_assist_club = SerieA_assists_page_soup.findAll('span',{"class":"wisbb_tableAbbrevLink"})
print(SerieA_assist_club[8].text)
print(SerieA_assist_club[1].text)
print("asd")
```

```python
SerieA_assist_club_list =[]
for i in range(50):
    get_club_name = SerieA_assist_club[i].text
    #print( get_club_name)
    get_club_name = get_club_name[1:4]
    SerieA_assist_club_list.append(get_club_name)
    
    #SerieA_assist_club_list.append(get_club_prefix)
    
print(SerieA_assist_club_list)
```

```python
PlayerID_SerieA_assist_list = []
for i in range(50):
    make_id = assist_SerieA_Name[i][0:4] + SerieA_assist_club_list[i]
    PlayerID_SerieA_assist_list.append(make_id)
    
print(PlayerID_SerieA_assist_list)
```

```python
SerieA_assist_data = list(zip(assist_SerieA_Name,SerieA_assist_club_list,SerieA_assists_made_list,PlayerID_SerieA_assist_list))
SerieA_assist_df = pd.DataFrame(data = SerieA_assist_data,columns = ['Player Name','Club','Assists','PlayerID'])
SerieA_assist_df.head()
```
