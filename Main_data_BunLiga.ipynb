```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import urllib.request
```

```python
Fox_goal_URL = 'https://www.foxsports.com/soccer/stats?competition=4&season=20180&category=standard&sort=3'
```

```python
uClient = uReq(Fox_goal_URL)
page_html = uClient.read()
uClient.close()
```

```python
Fox_goal_page_soup = soup(page_html,"html.parser")
```

```python
Fox_goal_Name_BunLiga = Fox_goal_page_soup.findAll('a',{"class":"wisbb_fullPlayer"})
#print(Fox_goal_Name_BunLiga[0].span.text)
print(len(Fox_goal_Name_BunLiga))
```

```python
Fox_goal_name = []
for i in range(50):
    get_Goal_playerName_Fox = Fox_goal_Name_BunLiga[i].span.text.split(', ')
    #print(T_get_assist_playerName)
    #a,b = T_get_assist_playerName.split(',')
    #if len(b) == 0
    if len(get_Goal_playerName_Fox) == 2:
        a =get_Goal_playerName_Fox[1]
        b = get_Goal_playerName_Fox[0]
        c = a+' '+b
    else:
        b = get_Goal_playerName_Fox[0]
        a = ''
        c = b
    Fox_goal_name.append(c)

print(Fox_goal_name)
```
