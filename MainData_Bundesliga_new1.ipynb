```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import urllib.request
```

```python
Bundesliga_assists_url = "https://www.foxsports.com/soccer/stats?competition=4&season=20180&category=standard&sort=4"
```

```python
uClient = uReq(Bundesliga_assists_url)
Bundesliga_assists_url_page_html = uClient.read()
uClient.close()
```

```python
Bundesliga_assists_page_soup = soup(Bundesliga_assists_url_page_html,"html.parser")
```

```python
assist_Bundesliga_playerName = Bundesliga_assists_page_soup.findAll("a",{"class":"wisbb_fullPlayer"})
print(assist_Bundesliga_playerName[0].span.text)
#a,b = assist_Bundesliga_playerName[14].span.text.split(',')
```

```python
assist_Bundesliga_Name = []
for i in range(50):
    get_assist_playerName = assist_Bundesliga_playerName[i].span.text.split(', ')
    #print(T_get_assist_playerName)
    #a,b = T_get_assist_playerName.split(',')
    #if len(b) == 0
    if len(get_assist_playerName) == 2:
        a = get_assist_playerName[1]
        b = get_assist_playerName[0]
        c = a+' '+b
    else:
        b = get_assist_playerName[0]
        a = ''
        c = b
    assist_Bundesliga_Name.append(c)
    
print(assist_Bundesliga_Name)
    
    
```

```python
Bundesliga_assistsPlusGoal_made = Bundesliga_assists_page_soup.findAll('tr',{"class":"wisbb_fvStand"})
print(Bundesliga_assistsPlusGoal_made[1].contents[9].text)
```

```python
Bundesliga_assistsPlusGoal_made_list = []
for i in range(1,51):
    Bundesliga_assists = Bundesliga_assistsPlusGoal_made[i].contents[9].text
    Bundesliga_assistsPlusGoal_made_list.append(Bundesliga_assists)
    
print(Bundesliga_assistsPlusGoal_made_list)
```

```python
Bundesliga_assists_made = Bundesliga_assists_page_soup.findAll('td',{"class":"wisbb_priorityColumn wisbb_selected"})
print(len(Bundesliga_assists_made))
```

```python
Bundesliga_assists_made_list = []
for i in range(50):
    Bundesliga_assists = Bundesliga_assists_made[i].text
    Bundesliga_assists_made_list.append(Bundesliga_assists)
    
print(Bundesliga_assists_made_list)
```

```python
Bundesliga_assist_club = Bundesliga_assists_page_soup.findAll('span',{"class":"wisbb_tableAbbrevLink"})
print(Bundesliga_assist_club[8].text)
print(Bundesliga_assist_club[1].text)
print("asd")
```

```python
Bundesliga_assist_club_list =[]
for i in range(50):
    get_club_name = Bundesliga_assist_club[i].text
    #print( get_club_name)
    get_club_name = get_club_name[1:4]
    Bundesliga_assist_club_list.append(get_club_name)
    
    #Bundesliga_assist_club_list.append(get_club_prefix)
    
print(Bundesliga_assist_club_list)
```

```python
PlayerID_Bundesliga_assist_list = []
for i in range(50):
    make_id = assist_Bundesliga_Name[i][0:4] + Bundesliga_assist_club_list[i]
    PlayerID_Bundesliga_assist_list.append(make_id)
    
print(PlayerID_Bundesliga_assist_list)
```

```python
Bundesliga_assist_data = list(zip(assist_Bundesliga_Name,Bundesliga_assist_club_list,Bundesliga_assists_made_list,PlayerID_Bundesliga_assist_list))
Bundesliga_assist_df = pd.DataFrame(data = Bundesliga_assist_data,columns = ['Player Name','Club','Assists','PlayerID'])
Bundesliga_assist_df.head()
```

```python
Bundesliga_assist_df['Goals'] = Bundesliga_assistsPlusGoal_made_list
```
